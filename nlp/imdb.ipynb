```python
import tensorflow as tf

dataset = tf.keras.utils.get_file(
    fname="aclImdb.tar.gz", 
    origin="http://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz", 
    extract=True,
)
```

```python
import os
import glob
dn = os.path.dirname(dataset)
glob.glob("/root/.keras/datasets/aclImdb/train/pos/*")
```

```python
import os
import glob
import pandas as pd
def getdata(mid):
    dn = os.path.dirname(dataset)
    posfn = glob.glob(os.path.join(dn, "aclImdb", mid, "pos", "*"))
    negfn = glob.glob(os.path.join(dn, "aclImdb", mid, "neg", "*"))
    contents = []
    for fn in posfn + negfn:
        with open(fn, encoding="utf-8") as f:
            contents.append(f.read())
    df = pd.DataFrame({
        "content":contents,
        "sentiment":[1] * len(posfn) + [0] * len(negfn)
    })
    return df
```

```python
train_df = getdata("train")
test_df = getdata("test")
```

```python
test_df
```

```python
# 1.Tokenize I love you -> 1 2 3
# fit_on_text(texts)tokentextslist
from tensorflow.keras.preprocessing.text import Tokenizer
toke=Tokenizer(num_words=3000)
toke.fit_on_texts(train_df["content"]) 
```

```python
#
#toke.word_index
#toke.index_word
x_train_seq=toke.texts_to_sequences(train_df["content"])
x_test_seq=toke.texts_to_sequences(test_df["content"])
```

```python
pd.DataFrame(x_train_seq)
```

```python
# 2. padding  
from tensorflow.keras.preprocessing.sequence import pad_sequences
x_train_pad = pad_sequences(x_train_seq,maxlen=512)
x_test_pad = pad_sequences(x_test_seq,maxlen=512)
```

```python
pd.DataFrame(x_train_pad)
```

```python
from threading import active_count
#Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,GlobalAveragePooling1D,Dense

layers=[
  Embedding(3001,100,mask_zero=True,input_length=512),
  GlobalAveragePooling1D(),
  Dense(2,activation="softmax")
]
model=Sequential(layers)
model.summary()
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
model.compile(loss=SparseCategoricalCrossentropy(),
       optimizer="adam",
       metrics=["accuracy"])
```

```python
import numpy as np
y_train = np.array(train_df["sentiment"])
y_test = np.array(test_df["sentiment"])
```

```python
# batch_size: (5[/]-200[/])
# epochs: (60000) ()
# validation_split: (val loss, )
# (60000 * 0.9) / 200 -> 270
# verbose: 0(quiet) 1(full) 2(no progress bar)
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
callbacks = [
    ModelCheckpoint("imdb.h5", save_best_only=True),
    EarlyStopping(patience=5, restore_best_weights=True)
]
model.fit(x_train_pad,
     y_train,
     batch_size=200,
     epochs=50,
     validation_split=0.1,
     verbose=2,
     callbacks=callbacks)
```

```python
model.evaluate(x_test_pad, y_test)
```
